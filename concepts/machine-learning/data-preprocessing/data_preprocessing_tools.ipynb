# Data Preprocessing Tools

## Importing the libraries

```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

## Importing the dataset

```python
dataset = pd.read_csv('data.csv')
features = dataset.iloc[:, 0:-1].to_numpy()
dependents = dataset.iloc[:, -1].to_numpy()

print(dataset, features, dependents, sep="\n\n")
```

## Taking care of missing data

```python
from sklearn.impute import SimpleImputer

imputer = SimpleImputer(missing_values=np.nan, strategy='mean')
features[:, 1:3] = imputer.fit(features[:, 1:3]).transform(features[:, 1:3])

print(features)
```

## Encoding categorical data

### Encoding the Independent Variable

```python
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import OneHotEncoder

# This might need to be turned into a numpy array with np.array()
features = ColumnTransformer(transformers=[('encoder', OneHotEncoder(), [0])], remainder='passthrough').fit_transform(features)

print(features)
```

### Encoding the Dependent Variable

```python
from sklearn.preprocessing import LabelEncoder

dependents = LabelEncoder().fit_transform(dependents)

print(dependents)
```

## Splitting the dataset into the Training set and Test set

```python
from sklearn.model_selection import train_test_split

features_train, features_test, dependents_train, dependents_test = train_test_split(features, dependents, test_size=0.2, random_state=1)

print(features_train, features_test, dependents_train, dependents_test, sep="\n\n")
```

## Feature Scaling

```python
from sklearn.preprocessing import StandardScaler

# Must apply the same scaler to both the training and test set, so they use the same calculated
# mean and standard deviation that was fitted from the training set.
sc = StandardScaler()
features_train[:, 3:] = sc.fit_transform(features_train[:, 3:])
features_test[:, 3:] = sc.transform(features_test[:, 3:])

print(features_train, features_test, sep="\n\n")
```

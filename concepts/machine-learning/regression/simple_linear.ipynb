# Simple Linear Regression

```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

## Import dataset

```python
dataset = pd.read_csv('salary_data.csv')
features = dataset.iloc[:, 0:-1].to_numpy()
dependents = dataset.iloc[:, -1].to_numpy()

print(dataset, features, dependents, sep="\n\n")
```

## Split into training and test data

```python
from sklearn.model_selection import train_test_split

features_train, features_test, dependents_train, dependents_test = train_test_split(features, dependents, test_size = 0.2, random_state = 0)
```

## Train the model on the training data

```python
from sklearn.linear_model import LinearRegression

model = LinearRegression()
model.fit(features_train, dependents_train)
```

## Predict the test data results

```python
predictions = model.predict(features_test)
```

## Visualize the training data

```python
plt.title('Salary vs Experience (Training Data)')
plt.xlabel('Years of Experience')
plt.ylabel('Salary')
plt.scatter(features_train, dependents_train, color = 'red')
plt.plot(features_train, model.predict(features_train), color = 'blue')
```

## Visualize the test data

```python
plt.title('Salary vs Experience (Test Data)')
plt.xlabel('Years of Experience')
plt.ylabel('Salary')
plt.scatter(features_test, dependents_test, color = 'red')
plt.plot(features_test, predictions, color = 'blue')
```

## Salary for a person with a certain amount of experience

```python
print(model.predict(np.array([6]).reshape(1, 1)))
```

```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

## Import & Pre-process

### Import from File

```python
from sklearn.model_selection import train_test_split

dataset = pd.read_csv('position_salaries.csv')
features = dataset.iloc[:, 1:-1].to_numpy()
dependents = dataset.iloc[:, -1].to_numpy()

print(dataset, features, dependents, sep="\n\n")

features_train, features_test, dependents_train, dependents_test = train_test_split(features, dependents, test_size = 0.2, random_state = 0)
```

### Encode (Optional)

The `level` column in the dataset already encoded the `position` column so we don't need to do any encoding step here. That is also why we didn't include the first column when grabbing features.

### Standardize & Normalize (Optional)

### Split into Training & Test Data

```python
from sklearn.model_selection import train_test_split

features_train, features_test, dependents_train, dependents_test = train_test_split(features, dependents, test_size = 0.2, random_state = 0)
```

## Train Model & Predict Test Data

### Train Model

```python
# Simple Linear Regression
from sklearn.linear_model import LinearRegression

linear_model = LinearRegression().fit(features, dependents)
```

```python
# Polynomial Regression
from sklearn.preprocessing import PolynomialFeatures

poly_reg = PolynomialFeatures(degree = 4)
poly_features = poly_reg.fit_transform(features)
poly_model = LinearRegression().fit(poly_features, dependents)
```

### Predict Test Data

```python
linear_predictions = linear_model.predict(features)
poly_predictions = poly_model.predict(poly_features)
```

## Visualize Results

```python
# Simple Linear Regression Results
plt.title('Level vs Salary (Simple Linear Regression)')
plt.xlabel('Position Level')
plt.ylabel('Salary')
plt.scatter(features, dependents, color = 'red')
plt.plot(features, linear_predictions, color = 'blue')
# print(np.concatenate((linear_predictions.reshape(len(linear_predictions), 1), dependents.reshape(len(dependents), 1)), axis=1))
```

```python
# Polynomial Regression Results
plt.title('Level vs Salary (Polynomial Linear Regression)')
plt.xlabel('Position Level')
plt.ylabel('Salary')
plt.scatter(features, dependents, color = 'red')
plt.plot(features, poly_predictions, color = 'blue')
# print(np.concatenate((poly_predictions.reshape(len(poly_predictions), 1), dependents.reshape(len(dependents), 1)), axis=1))
```

### Predict New Results

```python
# Simple Linear Regression
print(linear_model.predict([[6.5]]))

# Polynomial Regression
print(poly_model.predict(poly_reg.fit_transform([[6.5]])))
```

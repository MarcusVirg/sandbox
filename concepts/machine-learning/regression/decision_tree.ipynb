```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

## Import & Pre-process

### Import from File

```python
dataset = pd.read_csv('position_salaries.csv')
features = dataset.iloc[:, 1:-1].to_numpy()
dependents = dataset.iloc[:, -1].to_numpy()

print(dataset, features, dependents, sep="\n\n")
```

## Train Model & Predict Test Data

### Train Model

```python
from sklearn.tree import DecisionTreeRegressor

# Passing random state fixes the random seed
model = DecisionTreeRegressor(random_state=0)

model.fit(features, dependents)
```

### Predict

```python
predictions = model.predict(features)
```

## Visualize Results

```python
plt.title('Level vs Salary (Decision Tree Regression)')
plt.xlabel('Position Level')
plt.ylabel('Salary')

x_grid = np.arange(min(features), max(features), 0.1)
x_grid = x_grid.reshape((len(x_grid), 1))
plt.scatter(features, dependents, color = 'red')
plt.plot(x_grid, model.predict(x_grid), color = 'blue')
```
